# BitTorrent Simulation Demo

This notebook exercises the modular simulation package, captures a few representative states, and prints compact summaries for inspection.

```python
from pathlib import Path

from bittorrent_sim import BitTorrentSimulation, SimulationConfig
from bittorrent_sim.visualization import build_layout, save_snapshot

ARTIFACTS = Path("artifacts")
ARTIFACTS.mkdir(exist_ok=True)

config = SimulationConfig(
    size_of_file=20,
    initial_number_of_nodes=24,
    peer_optimistic_selection_count=2,
    total_number_of_connections=4,
    optimistically_unchoke_constant=4,
)
simulation = BitTorrentSimulation(config=config, seed=7)
pos = build_layout(simulation)
simulation.summary()
```

```python
initial_path = save_snapshot(simulation, ARTIFACTS / "snapshot-initial.png", pos=pos, title="Initial State")
initial_path
```

```python
for _ in range(6):
    simulation.update()

mid_path = save_snapshot(simulation, ARTIFACTS / "snapshot-after-updates.png", pos=pos, title="After Six Updates")
simulation.summary()
```

```python
added_node = simulation.add_node()
removed_node = simulation.delete_node()

churn_path = save_snapshot(simulation, ARTIFACTS / "snapshot-after-churn.png", pos=pos, title="After Add/Delete Churn")
{
    "added_node": added_node,
    "removed_node": removed_node,
    **simulation.summary(),
}
```
